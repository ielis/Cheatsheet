```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

"rc params file being used at the moment: '{}'".format(matplotlib.matplotlib_fname())
```

```python
x = np.linspace(0, 10)
y = np.sin(x)
fig, ax = plt.subplots()
line = ax.plot(x,y,'-')
ax.set(title="Sinus wave", xlabel=r'$x$', ylabel=r'$y = sin(x)$', xticks=list(range(11)));
```
